```python
import pandas as pd

# Chemin d'accès au fichier de données
bf_file_path = 'bf.txt'
```

```python
# Lecture du contenu du fichier
with open(bf_file_path, 'r', encoding='utf-8') as file:
    content = file.read()

# Affichage du contenu du fichier
print(content)
```

```python
# Colonnes à sélectionner dans le DataFrame
columns = ['geonameid', 'name', 'latitude', 'longitude']

# Lecture du fichier CSV dans un DataFrame pandas
df = pd.read_csv(bf_file_path, sep='\t', encoding='utf-8-sig', usecols=[0, 1, 4, 5], names=columns)
```

```python
# Renommage des colonnes
df.columns = ['ID', 'location_name', 'lat', 'long']
```

```python
# Sauvegarde des données dans un fichier CSV nommé burkina_location.csv
df.to_csv('burkina_location.csv', index=False,encoding='utf-8-sig')
```

```python
# Chargement du fichier CSV burkina_location.csv
df = pd.read_csv('burkina_location.csv')
```

```python
# Extraction des données pour le lieu 'gounghin'
gounghin_datas = df[df['location_name'].str.lower().str.contains('gounghin')]
gounghin_datas.to_csv('gounghin.csv', index=False)
```

```python
# Extraction des données pour les lieux dont le nom commence par une lettre entre A et P
subset_datas = df[df['location_name'].str[0].between('A', 'P', inclusive='both')]
subset_datas.to_csv('subset_datas.csv', index=False, encoding='utf-8-sig')
```

```python
# Identification des valeurs minimales de latitude et longitude
min_lat_data = df.loc[df['lat'].idxmin()]
min_lon_data = df.loc[df['long'].idxmin()]
min_lat = min_lat_data['lat']
min_lon = min_lon_data['long']
min_lat_location = min_lat_data['location_name']
min_lon_location = min_lon_data['location_name']
```

```python
# Filtrage des données basé sur des conditions spécifiques
filtered_data = df[(df['lat'] >= 11) & (df['long'] <= 0.5)]
```

```python
# Affichage des résultats concernant les valeurs minimales de latitude et longitude
print("------------------Coordonnées Minimales de Latitude et Longitude------------------")
print(f"Latitude minimale : {min_lat}, Lieu : {min_lat_location}.")
print(f"Longitude minimale : {min_lon}, Lieu : {min_lon_location}.")

# Affichage des lieux dont les coordonnées sont comprises entre latitude >= 11 et longitude <= 0.5
print("=================Lieux avec Coordonnées (lat >= 11 et lon <= 0.5) :=================")
print(filtered_data[['location_name', 'lat', 'long']])
```

```python
# Chargement des données extraites du lieu 'gounghin'
gounghin_datas = pd.read_csv('gounghin.csv')

# Chargement des données extraites pour les lieux commençant par A à P
subset_datas = pd.read_csv('subset_datas.csv')
```

```python
# Chemin d'accès au fichier Excel de sortie
excel_file_path = 'mini_projet.xlsx'

# Écriture des données dans un fichier Excel
with pd.ExcelWriter(excel_file_path, engine='xlsxwriter') as writer:
    # Sauvegarde des données du lieu 'gounghin' dans une feuille Excel
    gounghin_datas.to_excel(writer, sheet_name='gounghin', index=False)

    # Sauvegarde des données des lieux commençant par A à P dans une feuille Excel
    subset_datas.to_excel(writer, sheet_name='A_to_P', index=False)
```
